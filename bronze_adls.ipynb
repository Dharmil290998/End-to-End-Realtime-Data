```python
import pandas as pd

files = [
  {"file": "map_cities.json"},
  {"file": "map_cancellation_reasons.json"},
  {"file": "map_payment_methods.json"},
  {"file": "map_ride_statuses.json"},
  {"file": "map_vehicle_makes.json"},
  {"file": "map_vehicle_types.json"}
]



for file in files:
    url = f"https://realtimedataproject.blob.core.windows.net/project1/bronzelayer/{file['file']}?sp=r&st=2026-04-07T03:16:13Z&se=2026-04-11T11:31:13Z&spr=https&sv=2024-11-04&sr=c&sig=p3QI8Ei%2BMVXpS%2B5YglZ5bfQsrQbiDNsu4Vmwgfgn9Es%3D"

    df = pd.read_json(url)
    df_spark = spark.createDataFrame(df)

    # Clean table name
    table_name = file['file'].replace(".json", "")

    # Write to Bronze
    df_spark.write.format("delta") \
        .mode("overwrite") \
        .option("overwriteSchema","true") \
        .saveAsTable(f"realtimedata.bronze.{table_name}")
```

```python
import pandas as pd

url = f"https://realtimedataproject.blob.core.windows.net/project1/bronzelayer/bulk_rides.json?sp=r&st=2026-04-07T03:16:13Z&se=2026-04-11T11:31:13Z&spr=https&sv=2024-11-04&sr=c&sig=p3QI8Ei%2BMVXpS%2B5YglZ5bfQsrQbiDNsu4Vmwgfgn9Es%3D"

df = pd.read_json(url)

df_spark = spark.createDataFrame(df)

if not spark.catalog.tableExists("uber.bronze.bulk_rides"):
    df_spark.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable("realtimedata.bronze.bulk_rides")

print("This will not run more than 1 time")
```
